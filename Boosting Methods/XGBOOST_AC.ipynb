```python
import pandas as pd
from keras.utils import to_categorical
import numpy as np
from sklearn.model_selection import train_test_split
pd.set_option('display.max_columns', 50, 'display.max_rows', 100)



c = pd.read_csv('australian.dat', header=None, sep=' ')
```

```python
X = c.loc[:,0:13]
y = c.loc[:,14]
X = pd.get_dummies(X, columns=[0,3,4,5,7,8,10,11])
#y = pd.get_dummies(y)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)
```

```python
from xgboost import XGBClassifier
from xgboost import plot_importance
import matplotlib.pyplot as plt

my_model = XGBClassifier(n_estimators=100)

my_model.fit(X_train, y_train)
preds = my_model.predict(X_test)
acc_gnb = (preds == y_test).sum().astype(float) / len(preds)*100
acc_gnb
```

```python
fig, ax = plt.subplots(figsize=(12 ,12))
plot_importance(my_model, ax=ax)
plt.show()
```
