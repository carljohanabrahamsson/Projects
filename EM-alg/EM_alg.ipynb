```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
from skimage.transform import rescale, resize
import imageio
from sklearn.mixture import GaussianMixture
from scipy.stats import norm



def read_image(name):
    image = imageio.imread(name)
    image = resize(image,(128,128))
    image = 255 * image
    image = image.astype(np.uint8)
    data = image.ravel()
    return image, data

def crop_image(image):
    image = image[100:120, 80:120]
    row_size = image.shape[0]
    col_size = image.shape[1]
    image = resize(image,(image.shape[0],image.shape[1]))
    image = 255 * image
    image = image.astype(np.uint8)
    cropped_data = image.ravel()
    return image, cropped_data, row_size, col_size

def show_image(image, row_size, col_size):
    image = np.array([image])
    image = np.reshape(image, (row_size,col_size))
    image = resize(image,(row_size,col_size))
    plt.imshow(image,cmap=cm.gray)
    plt.show()

def add_noise(data, sigma2):
    mean = 0
    gauss = np.random.normal(mean, sigma2, len(data))
    data = data + gauss
    data[data > 255] = 255
    data[data < 0] = 0
    return data

def compute_prob_x(image):
    histo = np.histogram(image.flatten(), range=(0, 255), bins = 256)
    #histo[0][histo[0] < 1] = 1
    prob_x = histo[0]/np.sum(histo[0])
    plt.plot(prob_x)
    plt.show()
    return prob_x

class EM_alg:
    
    def __init__(self, *args, **kwargs):
        cropped_data.__init__(self, *args, **kwargs)
        prob_x.__init__(self, *args, **kwargs)
        sigma2.__init__(self, *args, **kwargs)
        K.__init__(self, *args, **kwargs)
        self.K = K
        data = cropped_data
        self.data = data
        N = len(self.data)
        self.N = N
        mu = np.random.randint(0, 255, K)
        Sigma = np.random.randint(1, sigma2, K)
        pi = np.random.rand(K)
        pi = pi/sum(pi)
        self.mu = mu
        self.Sigma = Sigma
        self.pi = pi
        self.E_step()
        
    def get_param(self):
        return self.mu, self.Sigma
        
    def E_step(self):
        gamma = np.zeros((self.K, self.N))
        for n in range(self.N):
            for k in range(self.K):
                gamma[k,n] = self.pi[k]*norm.pdf(self.data[n], self.mu[k], self.Sigma[k])
        gamma /= gamma.sum(0)
        self.gamma = gamma
        self.M_step()
    
    def M_step(self):
        N_k = np.zeros((self.K))
        mu_old = np.copy(self.mu)
        Sigma_old = np.copy(self.Sigma)
        pi_old = np.copy(self.pi)
        for k in range(self.K):
            N_k[k] = np.sum(self.gamma[k,:])
            self.mu[k] = 0
            for n in range(self.N):
                self.mu[k] += self.gamma[k,n]*self.data[n]
            self.mu[k] /= N_k[k]
            self.pi[k] = N_k[k]/self.N
            self.Sigma[k] = 0
            for n in range(self.N):
                self.Sigma[k] += self.gamma[k,n]*(self.data[n]-self.mu[k])**2
            self.Sigma[k] /= N_k[k]
        self.Sigma[self.Sigma < 1] = 1
        self.Sigma = np.sqrt(self.Sigma)
        old_vals = np.zeros((len(self.mu), 3))
        old_vals[:,0] = np.copy(mu_old)
        old_vals[:,1] = np.copy(Sigma_old)
        old_vals[:,2] = np.copy(pi_old)
        new_vals = np.zeros((len(self.mu), 3))
        new_vals[:,0] = np.copy(self.mu)
        new_vals[:,1] = np.copy(self.Sigma)
        new_vals[:,2] = np.copy(self.pi)
        matrix_vals = abs(old_vals - new_vals)
        print(self.mu, self.Sigma, end="\r")
        if len(matrix_vals[matrix_vals > 0.1].ravel()) < 1:
            print(self.mu, self.Sigma, 'end')
        else:
            self.E_step()
            
def plot_gmm(prob_x, mu, sigma):
    plt.plot(prob_x)
    print(prob_x.sum())
    g = np.zeros((len(mu),len(prob_x)))
    x = np.arange(len(prob_x))
    for i in range(len(mu)):
        if mu[i] != 0:
            g[i,:] += np.exp(-np.power(x - mu[i], 2.) / (2 * np.power(sigma[i], 2.)))
            g[i,:] = g[i,:]/sum(g[i,:])
            #plt.plot(g[i,:])
            #plt.show()
    #p = p/sum(p)
    a = g.sum(axis=0)/sum(g.sum(axis=0))
    print(a.sum())
    plt.plot(a)
    plt.show()

if __name__=="__main__":
    name="lena_gray.jpg"
    image, data = read_image(name)
    cropped_image, cropped_data, row_size, col_size = crop_image(image)
    show_image(cropped_image, row_size, col_size)
    sigma2 = 20
    noisy_image = add_noise(cropped_data, sigma2)
    show_image(noisy_image, row_size, col_size)
    prob_x = compute_prob_x(cropped_image)
    K=3
    f = EM_alg(cropped_data, prob_x, sigma2, K)
    mu, sigma = f.get_param()
    plot_gmm(prob_x, mu, sigma)
```
